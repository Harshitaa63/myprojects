```python
import cv2
```

```python
import matplotlib.pyplot as plt
#read the image
image= cv2.imread('/content/image.jpg')
#show the image
plt.imshow(image)
```

```python
#convert to grey scale
grey_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#show the image
plt.imshow(grey_image, cmap='gray')
```

```python
#create sift  (SIFT excelled at finding scale-invariant blobs across cells and their nuclei.)
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(grey_image, None)
```

```python
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None,
                                         flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS,
                                         color=(0, 255, 0))
```

```python
# Display Keypoints
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(image_with_keypoints, cv2.COLOR_BGR2RGB))
plt.title("SIFT Feature Keypoints")
plt.axis('off')
plt.show()
```

```python
# Initialize ORB Detector (ORB efficiently marked corners and boundaries where intensity shifts were strongest.)
orb = cv2.ORB_create(nfeatures=1000)
```

```python
#Detect Keypoints and Compute Descriptors
keypoints, descriptors = orb.detectAndCompute(grey_image, None)
```

```python
# Draw Keypoints on Image
keypoint_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)
```

```python
# Display Image
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(keypoint_image, cv2.COLOR_BGR2RGB))
plt.title("ORB Keypoints")
plt.axis("off")
plt.show()
```

```python
# count keypoints
print(f"Number of keypoints detected: {len(keypoints)}")
print(f"Descriptor shape: {descriptors.shape}")
```

```python
# initialize Brisk (quickly identified tight corners around the stained cell shapes.)
brisk = cv2.BRISK_create()
```

```python
#Detect Keypoints and Compute Descriptors
keypoints, descriptors = orb.detectAndCompute(grey_image, None)
keypoints, descriptors = brisk.detectAndCompute(grey_image, None)
```

```python
# draw kwypoints
keypoint_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)
```

```python
#display image
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(keypoint_image, cv2.COLOR_BGR2RGB))
plt.title("BRISK Keypoints")
plt.axis("off")
plt.show()
```

```python
# count key points
print(f"Number of keypoints detected: {len(keypoints)}")
print(f"Descriptor shape: {descriptors.shape}")
```

```python
# Initialize AKAZE Detector (AKAZE highlighted internal cell structures, leveraging the contrast created by staining.)
akaze = cv2.AKAZE_create()
```

```python
# dedect keypoints
keypoints, descriptors = akaze.detectAndCompute(grey_image, None)
```

```python
# draw key points
keypoint_image = cv2.drawKeypoints(image, keypoints, None, color=(0, 255, 0), flags=0)
```

```python
# display the image
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(keypoint_image, cv2.COLOR_BGR2RGB))
plt.title("AKAZE Keypoints")
plt.axis("off")
plt.show()
```
